#### Clase automovil con 3 atributos funciones leerdatos(), mostrardatos()

#### crear 3 objetos carro1, carro2, carro3

```python
class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca del automovil: ")
        self.placa = input("Ingrese la placa del automovil: ")
        self.precio = input("Ingrese el precio del automovil: ")
        return self.marca, self.placa, self.precio

    def mostrardatos(self):
        print("Marca:", self.marca)
        print("Placa:", self.placa)
        print("Precio:", self.precio)
        return self.marca, self.placa, self.precio
# fin de la clase automovil


print("Carrro Numero 1")
carro1 = Automovil("", "", "") # craeando el objeto carro1
carro1.leerdatos()
carro1.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*")
print("Carrro Numero 2")
carro2 = Automovil("", "", "") # craeando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("-*-*-*-*-*-*-*-*-*-*-*")
print("Carrro Numero 3")
carro3 = Automovil("", "", "") # craeando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print("sofware desarrollado por Eliseo Nava NL = 39")
```
